```python
# Sort the data by tenure in descending order and show the top 5 rows

df.sort_values("tenure", ascending=False).head()
```

```python
# Sort the data by MonthlyCharges in descending order and show the top 5 rows

df.sort_values("MonthlyCharges", ascending=False).head()
```

```python
# Create a new column called "HighMonthlyCharge" that is True if the MonthlyCharges is greater than 70 and False otherwise

df["HighMonthlyCharge"] = df["MonthlyCharges"] > 70
df[["MonthlyCharges", "HighMonthlyCharge"]].head(10)
```

```python
df["TenureGroup"] = df["tenure"].apply(lambda x: "New" if x < 12 else "Long-term")
df[["tenure", "TenureGroup"]].head()
```

```python
df.groupby("Churn")["MonthlyCharges"].mean()
```

```python
df.groupby("TenureGroup")["Churn"].value_counts()
```
