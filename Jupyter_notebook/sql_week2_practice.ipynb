### Calculate Churn Percentage by Contract

```python
#

query = """
SELECT 
    Contract,
    COUNT(*) AS total_customers,
    SUM(CASE WHEN Churn = 'Yes' THEN 1 ELSE 0 END) AS churned_customers,
    ROUND(
        100.0 * SUM(CASE WHEN Churn = 'Yes' THEN 1 ELSE 0 END) / COUNT(*),
        2
    ) AS churn_rate_percent
FROM churn_data
GROUP BY Contract
ORDER BY churn_rate_percent DESC
"""
pd.read_sql(query, conn)
```

#### This query shows:

- total customers in each contract type
- how many churned
- the churn rate as a percentage

This is much more useful than count alone.

## Week 2 Day 4
### Topic: HAVING, deeper grouped analysis, and ranked business insights

#### Today you will learn a very useful SQL idea:

- WHERE filters rows before grouping
- HAVING filters groups after grouping

#### This is important when you want to say things like:

- show only groups with more than 500 customers
- show only categories with churn rate above 40%
- show only payment methods with high average monthly charges

### Day 4 Goals
- understand the difference between WHERE and HAVING
- use HAVING with COUNT() and AVG()
- filter grouped results
- identify important business segments
- write clearer SQL findings

---

```python
# Import the library of pandas and sqlite3

import pandas as pd
import sqlite3  
df = pd.read_csv("../Data/Processed/telco_churn_cleaned.csv")
conn = sqlite3.connect(":memory:")
df.to_sql("churn_data", conn, index=False, if_exists="replace") 
```

```python
# This only keep contract group whose size is greater than 1000

query = """
SELECT Contract, COUNT(*) AS total_customers
FROM churn_data
GROUP BY Contract
HAVING COUNT(*) > 1000
ORDER BY total_customers DESC
"""
pd.read_sql(query, conn)
```

```python
# Filter the group by average monthly charge greater than 70

query = """
SELECT PaymentMethod, AVG(MonthlyCharges) AS avg_monthly_charge
FROM churn_data
GROUP BY PaymentMethod
HAVING AVG(MonthlyCharges) > 70
ORDER BY avg_monthly_charge DESC
"""
pd.read_sql(query, conn)
```

```python
# Show contract groups with more than 500 churned customers

query = """
SELECT Contract, COUNT(*) AS churned_customers
FROM churn_data
WHERE Churn = 'Yes'
GROUP BY Contract
HAVING COUNT(*) > 500
ORDER BY churned_customers DESC
"""
pd.read_sql(query, conn)
```

```python
# Show contract groups with more than 20% churn rate

query = """
SELECT 
    Contract,
    COUNT(*) AS total_customers,
    SUM(CASE WHEN Churn = 'Yes' THEN 1 ELSE 0 END) AS churned_customers,
    ROUND(
        100.0 * SUM(CASE WHEN Churn = 'Yes' THEN 1 ELSE 0 END) / COUNT(*),
        2
    ) AS churn_rate_percent
FROM churn_data
GROUP BY Contract
HAVING ROUND(
        100.0 * SUM(CASE WHEN Churn = 'Yes' THEN 1 ELSE 0 END) / COUNT(*),
        2
    ) > 20
ORDER BY churn_rate_percent DESC
"""
pd.read_sql(query, conn)
```

```python
# Payment methods with churn rate above 30%
query = """
SELECT 
    PaymentMethod,
    COUNT(*) AS total_customers,
    SUM(CASE WHEN Churn = 'Yes' THEN 1 ELSE 0 END) AS churned_customers,
    ROUND(
        100.0 * SUM(CASE WHEN Churn = 'Yes' THEN 1 ELSE 0 END) / COUNT(*),
        2
    ) AS churn_rate_percent
FROM churn_data
GROUP BY PaymentMethod
HAVING ROUND(
        100.0 * SUM(CASE WHEN Churn = 'Yes' THEN 1 ELSE 0 END) / COUNT(*),
        2
    ) > 30
ORDER BY churn_rate_percent DESC
"""
pd.read_sql(query, conn)
```

```python
# Internet service groups with average monthly charges above 60

query = """
SELECT InternetService, AVG(MonthlyCharges) AS avg_monthly_charge
FROM churn_data
GROUP BY InternetService
HAVING AVG(MonthlyCharges) > 60
ORDER BY avg_monthly_charge DESC
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT 
    InternetService,
    COUNT(*) AS total_customers,
    AVG(MonthlyCharges) AS avg_monthly_charge
FROM churn_data
GROUP BY InternetService
HAVING AVG(MonthlyCharges) > 60
ORDER BY avg_monthly_charge DESC
"""
pd.read_sql(query, conn)
```

## Day 4 SQL Summary

Today, I learned how to use `HAVING` in SQL to filter grouped results. 
Unlike `WHERE`, which filters rows before grouping, `HAVING` filters groups after aggregation. 
I used `HAVING` with `COUNT()`, `AVG()`, and churn rate calculations to identify high-risk customer groups.
