### Calculate Churn Percentage by Contract

```python
#

query = """
SELECT 
    Contract,
    COUNT(*) AS total_customers,
    SUM(CASE WHEN Churn = 'Yes' THEN 1 ELSE 0 END) AS churned_customers,
    ROUND(
        100.0 * SUM(CASE WHEN Churn = 'Yes' THEN 1 ELSE 0 END) / COUNT(*),
        2
    ) AS churn_rate_percent
FROM churn_data
GROUP BY Contract
ORDER BY churn_rate_percent DESC
"""
pd.read_sql(query, conn)
```

#### This query shows:

- total customers in each contract type
- how many churned
- the churn rate as a percentage

This is much more useful than count alone.

```python
# Import the library of pandas and sqlite3

import pandas as pd
import sqlite3  
df = pd.read_csv("../Data/Processed/telco_churn_cleaned.csv")
conn = sqlite3.connect(":memory:")
df.to_sql("churn_data", conn, index=False, if_exists="replace") 
```

```python
# This only keep contract group whose size is greater than 1000

query = """
SELECT Contract, COUNT(*) AS total_customers
FROM churn_data
GROUP BY Contract
HAVING COUNT(*) > 1000
ORDER BY total_customers DESC
"""
pd.read_sql(query, conn)
```

```python
# Filter the group by average monthly charge greater than 70

query = """
SELECT PaymentMethod, AVG(MonthlyCharges) AS avg_monthly_charge
FROM churn_data
GROUP BY PaymentMethod
HAVING AVG(MonthlyCharges) > 70
ORDER BY avg_monthly_charge DESC
"""
pd.read_sql(query, conn)
```

```python
# Show contract groups with more than 500 churned customers

query = """
SELECT Contract, COUNT(*) AS churned_customers
FROM churn_data
WHERE Churn = 'Yes'
GROUP BY Contract
HAVING COUNT(*) > 500
ORDER BY churned_customers DESC
"""
pd.read_sql(query, conn)
```

```python
# Show contract groups with more than 20% churn rate

query = """
SELECT 
    Contract,
    COUNT(*) AS total_customers,
    SUM(CASE WHEN Churn = 'Yes' THEN 1 ELSE 0 END) AS churned_customers,
    ROUND(
        100.0 * SUM(CASE WHEN Churn = 'Yes' THEN 1 ELSE 0 END) / COUNT(*),
        2
    ) AS churn_rate_percent
FROM churn_data
GROUP BY Contract
HAVING ROUND(
        100.0 * SUM(CASE WHEN Churn = 'Yes' THEN 1 ELSE 0 END) / COUNT(*),
        2
    ) > 20
ORDER BY churn_rate_percent DESC
"""
pd.read_sql(query, conn)
```

```python
# Payment methods with churn rate above 30%
query = """
SELECT 
    PaymentMethod,
    COUNT(*) AS total_customers,
    SUM(CASE WHEN Churn = 'Yes' THEN 1 ELSE 0 END) AS churned_customers,
    ROUND(
        100.0 * SUM(CASE WHEN Churn = 'Yes' THEN 1 ELSE 0 END) / COUNT(*),
        2
    ) AS churn_rate_percent
FROM churn_data
GROUP BY PaymentMethod
HAVING ROUND(
        100.0 * SUM(CASE WHEN Churn = 'Yes' THEN 1 ELSE 0 END) / COUNT(*),
        2
    ) > 30
ORDER BY churn_rate_percent DESC
"""
pd.read_sql(query, conn)
```

```python
# Internet service groups with average monthly charges above 60

query = """
SELECT InternetService, AVG(MonthlyCharges) AS avg_monthly_charge
FROM churn_data
GROUP BY InternetService
HAVING AVG(MonthlyCharges) > 60
ORDER BY avg_monthly_charge DESC
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT 
    InternetService,
    COUNT(*) AS total_customers,
    AVG(MonthlyCharges) AS avg_monthly_charge
FROM churn_data
GROUP BY InternetService
HAVING AVG(MonthlyCharges) > 60
ORDER BY avg_monthly_charge DESC
"""
pd.read_sql(query, conn)
```

```python
# import the library and read the data

import pandas as pd
import sqlite3

df = pd.read_csv("../Data/Processed/telco_churn_cleaned.csv")
conn = sqlite3.connect(":memory:")
```

```python
# Make the customer table

customer_df = df[[
    "customerID", "gender", "SeniorCitizen", "Partner", "Dependents",
    "tenure", "Churn"
]]
```

```python
# Make a service table

service_df = df[[
    "customerID", "PhoneService", "MultipleLines", "InternetService",
    "OnlineSecurity", "OnlineBackup", "DeviceProtection",
    "TechSupport", "StreamingTV", "StreamingMovies"
]]
```

```python
# Make a billing table

billing_df = df[[
    "customerID", "Contract", "PaperlessBilling", "PaymentMethod",
    "MonthlyCharges", "TotalCharges"
]]
```

```python
# Save These Tables into SQLite
# Right now we have three tables in our SQLite database:SQLite database: customer_data, service_data, and billing_data. 


customer_df.to_sql("customer_data", conn, index=False, if_exists="replace")
service_df.to_sql("service_data", conn, index=False, if_exists="replace")
billing_df.to_sql("billing_data", conn, index=False, if_exists="replace")
```

A JOIN combines rows from two or more tables using a common column.

In this case, the common column is 'customerID'

```python
query = """
SELECT 
    c.customerID,
    c.tenure,
    c.Churn,
    b.Contract,
    b.MonthlyCharges
FROM customer_data c
INNER JOIN billing_data b
    ON c.customerID = b.customerID
LIMIT 10
"""
pd.read_sql(query, conn)
```

### What does this mean?
- customer_data c → use c as short name for customer table
- billing_data b → use b as short name for billing table
- INNER JOIN → only keep rows where customerID exists in both tables
- ON c.customerID = b.customerID → tells SQL how to match the rows

```python
# Add Service Data with Another JOIN, I combine all 3 tables together

query = """
SELECT 
    c.customerID,
    c.tenure,
    c.Churn,
    b.Contract,
    b.MonthlyCharges,
    s.InternetService,
    s.TechSupport
FROM customer_data c
INNER JOIN billing_data b
    ON c.customerID = b.customerID
INNER JOIN service_data s
    ON c.customerID = s.customerID
LIMIT 10
"""
pd.read_sql(query, conn)
```

As you can see, we use In the query above:

- c = customer_data
- b = billing_data
- s = service_data

This makes queries shorter and easier to read.

```python
# Business Question with JOIN
# Question: Which contract types have the highest churn rate?

query = """
SELECT 
    b.Contract,
    COUNT(*) AS total_customers,
    SUM(CASE WHEN c.Churn = 'Yes' THEN 1 ELSE 0 END) AS churned_customers,
    ROUND(
        100.0 * SUM(CASE WHEN c.Churn = 'Yes' THEN 1 ELSE 0 END) / COUNT(*),
        2
    ) AS churn_rate_percent
FROM customer_data c
INNER JOIN billing_data b
    ON c.customerID = b.customerID
GROUP BY b.Contract
ORDER BY churn_rate_percent DESC
"""
pd.read_sql(query, conn)
```

```python
# Question: Which internet service type has the highest churn rate?

query = """
SELECT 
    s.InternetService,
    COUNT(*) AS total_customers,
    SUM(CASE WHEN c.Churn = 'Yes' THEN 1 ELSE 0 END) AS churned_customers,
    ROUND(
        100.0 * SUM(CASE WHEN c.Churn = 'Yes' THEN 1 ELSE 0 END) / COUNT(*),
        2
    ) AS churn_rate_percent
FROM customer_data c
INNER JOIN service_data s
    ON c.customerID = s.customerID
GROUP BY s.InternetService
ORDER BY churn_rate_percent DESC
"""
pd.read_sql(query, conn)
```

```python
# Question: What is the average monthly charge for each internet service type?

query = """
SELECT 
    s.InternetService,
    AVG(b.MonthlyCharges) AS avg_monthly_charge
FROM service_data s
INNER JOIN billing_data b
    ON s.customerID = b.customerID
GROUP BY s.InternetService
ORDER BY avg_monthly_charge DESC
"""
pd.read_sql(query, conn)
```

### Understand LEFT JOIN
It is useful when:

- you want to keep all customers
- some customers may not have matching service or billing records
- you want to check for missing matches

In your current cleaned dataset, almost everything should match,

so INNER JOIN and LEFT JOIN may look similar.

But in real databases, they can be very different.

```python
query = """
SELECT 
    c.customerID,
    c.Churn,
    b.Contract
FROM customer_data c
LEFT JOIN billing_data b
    ON c.customerID = b.customerID
LIMIT 10
"""
pd.read_sql(query, conn)
```
