```python
# Load the data

import pandas as pd
import numpy as np

df = pd.read_csv("../Data/Processed/telco_churn_cleaned.csv")
df.head()
```

```python
# Choose the Columns 

df[["tenure", "MonthlyCharges", "TotalCharges"]].head()
```

```python
df[["tenure", "MonthlyCharges", "TotalCharges"]].describe()
```

```python
df["MonthlyCharges"].mean()
df["MonthlyCharges"].median()
df["MonthlyCharges"].min()
df["MonthlyCharges"].max()
df["MonthlyCharges"].std()
```

```python
df["tenure"].mean()
df["tenure"].median()
df["tenure"].min()
df["tenure"].max()
df["tenure"].std()
```

```python
# Calculate Range

monthly_range = df["MonthlyCharges"].max() - df["MonthlyCharges"].min()
tenure_range = df["tenure"].max() - df["tenure"].min()
total_range = df["TotalCharges"].max() - df["TotalCharges"].min()

print("MonthlyCharges range:", monthly_range)
print("Tenure range:", tenure_range)
print("TotalCharges range:", total_range)
```

#### Compare Mean and Median

- mean is close to median means distribution may be more balanced
- mean is much higher or lower than median means distribution may be skewed

```python
stats_df = pd.DataFrame({
    "Column": ["tenure", "MonthlyCharges", "TotalCharges"],
    "Mean": [
        df["tenure"].mean(),
        df["MonthlyCharges"].mean(),
        df["TotalCharges"].mean()
    ],
    "Median": [
        df["tenure"].median(),
        df["MonthlyCharges"].median(),
        df["TotalCharges"].median()
    ]
})

stats_df
```

```python
df.groupby("Churn")[["tenure", "MonthlyCharges", "TotalCharges"]].mean()
```

```python
df.groupby("Churn")[["tenure", "MonthlyCharges", "TotalCharges"]].median()
```

```python
df[["tenure", "MonthlyCharges", "TotalCharges"]].agg(["mean", "median", "min", "max", "std"])
```

```python
summary_df = pd.DataFrame({
    "Feature": ["tenure", "MonthlyCharges", "TotalCharges"],
    "Mean": [
        df["tenure"].mean(),
        df["MonthlyCharges"].mean(),
        df["TotalCharges"].mean()
    ],
    "Median": [
        df["tenure"].median(),
        df["MonthlyCharges"].median(),
        df["TotalCharges"].median()
    ],
    "Min": [
        df["tenure"].min(),
        df["MonthlyCharges"].min(),
        df["TotalCharges"].min()
    ],
    "Max": [
        df["tenure"].max(),
        df["MonthlyCharges"].max(),
        df["TotalCharges"].max()
    ],
    "Std": [
        df["tenure"].std(),
        df["MonthlyCharges"].std(),
        df["TotalCharges"].std()
    ]
})

summary_df
```
