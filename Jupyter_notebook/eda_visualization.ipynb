```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/processed/telco_churn_cleaned.csv")
df.head()
```

#### Part 2: Visualize the Target Variable

```python
df["Churn"].value_counts()
```

```python
# This shows the distribution of customers who have churned vs. those who have not.

sns.countplot(x="Churn", data=df)
plt.title("Distribution of Churn")
plt.xlabel("Churn")
plt.ylabel("Count")
plt.show()
```

```python
# This shows how long customers have stayed with the company.

sns.histplot(df["tenure"], kde=True)
plt.title("Distribution of Tenure")
plt.xlabel("Tenure (months)")
plt.ylabel("Frequency")
plt.show()
```

```python
# This shows the distribution of monthly charges for customers.
sns.histplot(df["MonthlyCharges"], kde=True)
plt.title("Distribution of Monthly Charges")
plt.xlabel("Monthly Charges")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(df["TotalCharges"], kde=True)
plt.title("Distribution of Total Charges")
plt.xlabel("Total Charges")
plt.ylabel("Frequency")
plt.show()
```

```python
# This shows the distribution of total charges for customers    
sns.histplot(df["TotalCharges"], kde=True)
plt.title("Distribution of Total Charges")
plt.xlabel("Total Charges")
plt.ylabel("Frequency")
plt.show()
```

### Compare Numerical Features with Churn
Now we want to see whether churned and non-churned customers look different.

```python
# This shows the relationship between tenure and churn.

sns.boxplot(x="Churn", y="tenure", data=df)
plt.title("Tenure by Churn")
plt.show()
```

```python
# This shows the relationship between monthly charges and churn.

sns.boxplot(x="Churn", y="MonthlyCharges", data=df)
plt.title("Monthly Charges by Churn")
plt.show()
```

```python
sns.boxplot(x="Churn", y="TotalCharges", data=df)
plt.title("Total Charges by Churn")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create 1 row and 3 columns
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# First graph: tenure
sns.boxplot(x="Churn", y="tenure", data=df, ax=axes[0])
axes[0].set_title("Tenure by Churn")

# Second graph: MonthlyCharges
sns.boxplot(x="Churn", y="MonthlyCharges", data=df, ax=axes[1])
axes[1].set_title("Monthly Charges by Churn")

# Third graph: TotalCharges
sns.boxplot(x="Churn", y="TotalCharges", data=df, ax=axes[2])
axes[2].set_title("Total Charges by Churn")

# Make layout cleaner
plt.tight_layout()

# Show all graphs together
plt.show()
```

## Initial EDA Findings

- The target variable is imbalanced, with more non-churn customers than churn customers.
- Customers with shorter tenure appear to be more likely to churn.
- Monthly charges may be related to churn behavior.
- Customers with month-to-month contracts seem more likely to churn than those with one-year or two-year contracts.
- Some payment methods may be associated with higher churn rates.

#### The result of compare Numerical Features with Churn
- First, customers who did not churn generally have a much longer tenure. Many non-churn customers have stayed with the company for more than 40 months, while customers who churned usually have a much shorter tenure, with the median around 10 months. This suggests that customers are more likely to leave during the early stage of their contract, while long-term customers are more stable and less likely to switch to another company.

- Second, customers who churned tend to have higher monthly charges. For non-churn customers, the median monthly charge is around $65, while for churn customers, the median monthly charge is higher. This indicates that higher monthly charges may be associated with a higher risk of churn, possibly because customers are more sensitive to price when they feel the service is too expensive.

- Finally, total charges should be interpreted together with both monthly charges and tenure. Total charges depend on how much a customer pays each month and how long the customer has stayed with the company. Although churn customers often have higher monthly charges, their total charges are usually lower because they leave earlier. In contrast, non-churn customers may pay slightly less per month, but they stay longer, so the company earns more total revenue from them over time.
