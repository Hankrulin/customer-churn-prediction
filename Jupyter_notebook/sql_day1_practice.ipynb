Part 1: Read the dataset by using SQL statement & syntax

```python
# Import the library of pandas and sqlite3

import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("../Data/Processed/telco_churn_cleaned.csv")
df.head()
```

```python
# Create an in-memory SQLite database

conn = sqlite3.connect(":memory:")
```

```python
# Load the DataFrame into the SQLite database
df.to_sql("churn_data", conn, index=False, if_exists="replace")
```

```python
query = """
SELECT *
FROM churn_data
LIMIT 5
"""
pd.read_sql(query, conn)
```

```python
# Learn how to use SELECT syntax and select specific columns & limit the number of rows

query = """
SELECT customerID, tenure, MonthlyCharges, Churn
FROM churn_data
LIMIT 10
"""
pd.read_sql(query, conn)
```

```python
# How to use WHERE syntax

query = """
SELECT customerID, tenure, MonthlyCharges, Churn
FROM churn_data
WHERE Churn = 'Yes'
LIMIT 10
"""
pd.read_sql(query, conn)    
```

```python
# Customers with tenure greater than 12 by WHERE syntax

query = """
SELECT customerID, tenure, MonthlyCharges, Churn
FROM churn_data
WHERE tenure > 12
LIMIT 10
"""
pd.read_sql(query, conn)
```

```python
# Multiple conditions: Customers with tenure greater than 12 and Churn = 'Yes'

query = """
SELECT customerID, tenure, MonthlyCharges, Contract, Churn
FROM churn_data
WHERE tenure > 12 AND Churn = 'Yes'
LIMIT 10
"""
pd.read_sql(query, conn)
```

```python
#Learn ORDER BY syntax: Customers with the highest monthly charges

query = """
SELECT customerID, MonthlyCharges, Churn
FROM churn_data
ORDER BY MonthlyCharges DESC
LIMIT 10
"""
pd.read_sql(query, conn)
```

```python
# Learn ORDER BY syntax: Lowest tenure first

query = """
SELECT customerID, tenure, Churn
FROM churn_data
ORDER BY tenure ASC
LIMIT 10
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT customerID, Contract, PaymentMethod, Churn
FROM churn_data
LIMIT 10
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT Contract, Churn
FROM churn_data
WHERE Contract = 'Month-to-month'
LIMIT 10
"""
pd.read_sql(query, conn)
```

```python
# Show the top 10 customers with the highest TotalCharges

query = """
SELECT customerID, TotalCharges, Churn
FROM churn_data
ORDER BY TotalCharges DESC
LIMIT 10
"""
pd.read_sql(query, conn)
```

```python
#  Show 10 customers who use Fiber optic and have MonthlyCharges greater than 80

query = """
SELECT customerID, TotalCharges, MonthlyCharges, InternetService, Churn
FROM churn_data
WHERE InternetService = 'Fiber optic' AND MonthlyCharges > 80
LIMIT 10
"""
pd.read_sql(query, conn)
```

### Which customers are paying the most each month? 
---

```python
# Which customers are paying the most each month? 
# The answer is 7569-NMZYQ

query = """
SELECT customerID, TotalCharges, MonthlyCharges, InternetService, Churn
FROM churn_data
ORDER BY MonthlyCharges DESC
LIMIT 10
"""
pd.read_sql(query, conn)
```

### Are churned customers more likely to be on month-to-month contracts?

```python
# Are churned customers more likely to be on month-to-month contracts?

query = """
SELECT customerID, Contract, Churn
FROM churn_data
WHERE Churn = 'Yes' 
ORDER BY Contract
LIMIT 10
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT 
    Contract,
    COUNT(*) AS churned_customers,
    ROUND(COUNT(*) * 100.0 / (
        SELECT COUNT(*)
        FROM churn_data
        WHERE Churn = 'Yes'
    ), 2) AS percentage
FROM churn_data
WHERE Churn = 'Yes'
GROUP BY Contract
ORDER BY churned_customers DESC
"""

pd.read_sql(query, conn)
```

Yes, Most of the customers more likely to be on month-to-month contracts.

### Are high-paying fiber optic users more likely to churn?
---

```python
# Are high-paying fiber optic users more likely to churn?

query = """
SELECT 
    InternetService, MonthlyCharges, Churn
FROM churn_data
WHERE InternetService = 'Fiber optic'
  AND MonthlyCharges > 100
ORDER BY MonthlyCharges DESC
LIMIT 10
"""   

pd.read_sql(query, conn)
```

```python
query = """
SELECT 
    Churn,
    COUNT(*) AS customer_count,
    ROUND(
        COUNT(*) * 100.0 / (
            SELECT COUNT(*)
            FROM churn_data
            WHERE InternetService = 'Fiber optic'
              AND MonthlyCharges > 100
        ), 2
    ) AS percentage
FROM churn_data
WHERE InternetService = 'Fiber optic'
  AND MonthlyCharges > 100
GROUP BY Churn
ORDER BY percentage DESC
"""

pd.read_sql(query, conn)
```

### Which customers have stayed the longest?
---

```python
query = """
SELECT customerID, tenure, Churn
FROM churn_data
ORDER BY tenure DESC
LIMIT 10
"""
pd.read_sql(query, conn)    
```

The longest-staying customers have a tenure of 72 months. In the top 10 results, all customers have Churn = No, which means they are still with the company. This suggests that customers with longer tenure are less likely to churn.
